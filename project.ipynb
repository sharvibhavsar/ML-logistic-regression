Drop columns to remove multicollinearity.

```python
df_org.drop("Gender_Female", axis=1,inplace=True)
```

```python
df_org.drop("Married_No",axis=1,inplace=True)
df_org.drop("Education_Graduate",axis=1,inplace=True)
df_org.drop("Self_Employed_No",axis=1,inplace=True)
df_org.drop("Property_Area_Rural",axis=1,inplace=True)
df_org.drop("Loan_Status_N",axis=1,inplace=True)
df_org.drop("Dependents_3+",axis=1,inplace=True)
```

```python
df_org.dtypes
```

Convert data to standard format

Missing values handling for numerical data.

```python
df_org["LoanAmount"].fillna(df_org["LoanAmount"].median(), inplace=True)
df_org["Loan_Amount_Term"].fillna(df_org["Loan_Amount_Term"].median(), inplace=True)
df_org["Credit_History"].fillna(df_org["Credit_History"].mode()[0], inplace=True)
df_org[["LoanAmount","Loan_Amount_Term","Credit_History"]]
```

Check for null values in these columns.

```python
df_org["LoanAmount"].isnull().sum()
```

```python
df_org["Loan_Amount_Term"].isnull().sum()
```

Logistic model.

```python
x=df_org.iloc[:,2:15]
y=df_org.iloc[:,-1]
```

Train  model.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.7, random_state=5)
```

Model fitting and checking y_train and y_test.

```python
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
y_test
```

Accuracy score.

```python
from sklearn.metrics import accuracy_score, confusion_matrix
acc_sc=accuracy_score(y_test, y_pred)
acc_sc
```

Confusion matrix.

```python
cm=confusion_matrix(y_test,y_pred)
cm
```
